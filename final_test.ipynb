```python
def changeIndex(results):
    rta = []
    for i in range(len(results)):
        #print(results[i])
        if(results[i] == 0):
            rta.append(2)
        elif results[i] == 1:
            rta.append(0)
        else:
            rta.append(1)
    return rta
```
